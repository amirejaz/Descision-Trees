```python
import numpy as np
import pandas as pd
from matplotlib.pylab import rcParams
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

# Best Fitted Decision Tree

```python
df1 = pd.read_csv("Iris.csv")
```

```python
df1.head()
```

```python
df1.shape
```

```python
X = df1.iloc[:,1:5].values
```

```python
y = df1.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
rcParams['figure.figsize'] = 80, 50
plot_tree(clf)
```

# Overfitting Decision Tree

```python
df2 = pd.read_csv("Social_Network_Ads.csv")
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2['Gender'].replace({'Male':0, 'Female':1}, inplace=True)
```

```python
X = df2.iloc[:,1:4].values
```

```python
y = df2.iloc[:, -1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
clf2 = DecisionTreeClassifier()
```

```python
clf2.fit(X_train, y_train)
```

```python
y_pred = clf2.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
rcParams['figure.figsize'] = 80, 50
plot_tree(clf2)
```

This might be the case of overfitting

# Reduce overfitting

```python
clf3 = DecisionTreeClassifier(max_depth=3)
```

```python
clf3.fit(X_train, y_train)
```

```python
y_pred1 = clf3.predict(X_test)
```

```python
accuracy_score(y_test, y_pred1)
```

```python
rcParams['figure.figsize'] = 80, 50
plot_tree(clf3)
```
